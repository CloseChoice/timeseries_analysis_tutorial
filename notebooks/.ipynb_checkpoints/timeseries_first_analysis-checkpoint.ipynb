#### In this tutorial we want to study the different tools for analyzing time series. We use time series provided by https://datamarket.com/data/list/?q=provider%3Atsdl

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os
```

```python
os.listdir('../data')
```

read the data and don't use the footer

```python
sheep = pd.read_csv('../data/annual-sheep-population-1000s-in.csv').iloc[:-1,:]
```

```python
sheep.columns = ['Year', 'Sheep_pop']
```

```python
sheep.plot()
```

```python
from statsmodels.tsa.stattools import acf
from statsmodels.graphics.tsaplots import plot_acf
```

```python
acf(sheep['Sheep_pop'])
```

```python
plot_acf(sheep['Sheep_pop'])
```

```python
from statsmodels.tsa.stattools import pacf
from statsmodels.graphics.tsaplots import plot_pacf
```

```python
pacf(sheep['Sheep_pop'])
```

```python
sheep.columns
```

```python
plot_pacf(sheep['Sheep_pop'])
```

Let's make the timeseries stationary

```python
sheep_stat = np.diff(sheep['Sheep_pop'].astype(float))
```

```python
plot_acf(sheep_stat)
```

```python
plot_pacf(sheep_stat)
```
